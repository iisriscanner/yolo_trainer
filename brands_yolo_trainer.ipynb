```python
import gdown
import zipfile
from ultralytics import YOLO
from pathlib import Path
CWD = Path.cwd()
print(CWD)
```

```python
DATASET_NAME = ""
DATASET_PATH = CWD / DATASET_NAME
```

```python
file_id = ""
gdown.download(id=file_id, output=f"{DATASET_NAME}.zip")
with zipfile.ZipFile(CWD / f"{DATASET_NAME}.zip", 'r') as zip_ref:
    zip_ref.extractall(".")
```

```python
# Load a model
model = YOLO('yolov8x.pt')  # load a pretrained model (recommended for training)
# model.info()
```

```python
# Train the model
results = model.train(data=DATASET_PATH / "dataset.yaml", epochs=200, imgsz=640, project=DATASET_PATH)
```

```python
model = YOLO(DATASET_PATH / "train" / "last.pt")  # load a partially trained model
# Resume training
results = model.train(resume=True)
```

```python
# Load a model
model = YOLO(DATASET_PATH / "train" /"best.pt")
# Customize validation settings
validation_results = model.val(data=DATASET_PATH / "dataset.yaml", split="test")
```

```python
print(validation_results.box.map)    # map50-95
print(validation_results.box.map50) # map50
print(validation_results.box.map75) # map75
print(validation_results.box.maps)  # a list contains map50-95 of each category)
```
